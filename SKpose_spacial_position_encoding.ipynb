# Read csv files into dataframe_dictionnary

```python
import os
import pandas as pd
import gc

# Directory containing the CSV files
folder_path = 'D:\Th√®se_Toufik\code\BVH\PyMO-master\InHARD_13\Skeleton\positions_3D'

# Initialize an empty dictionary to store DataFrames
df_dict = {}

# Iterate over files in the directory
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        print(filename)
        # Read CSV file into a DataFrame
        df = pd.read_csv(os.path.join(folder_path, filename))
        
        # Add DataFrame to the dictionary with the filename (without extension) as key
        df_dict[os.path.splitext(filename)[0]] = df

# Now dataframes_dict contains all your DataFrames, with filenames as keys
```

# resample to 30 fps

```python
import pandas as pd

df_dict_30fps = {}
step = 4 # for 30fps: default 119fps 


# Define a function to create the new dataframe
def create_new_dataframe(old_df, interval):
    new_data = []
    for i in range(0, len(old_df), interval):
        # Take the last row in each interval
        new_data.append(old_df.iloc[i:i+interval].iloc[-1])
    # Create a new dataframe from the collected data
    new_dataframe = pd.DataFrame(new_data)
    return new_dataframe

# Iterate through the original dictionary and create new filtered DataFrames
for key, df in df_dict.items():
    print(key)
    # Create the new dataframe
    df_dict_30fps[key] = create_new_dataframe(df, step)
```

# Position encoding

### Generate position encoding

```python
import math
import numpy as np

def pn(x, N):
    """
    Generate positional encoding for a single coordinate (x, y, or z).

    Args:
    - x (float): The coordinate to be encoded (x, y, or z).
    - N (int): The dimension of the embedding, must be divisible by 2.

    Returns:
    - pos_encoding (np.ndarray): The positional encoding for the coordinate.
    """
    result = []
    for i in range(N + 1):
        result.append(np.float32(math.sin(2**i * math.pi * x)))
        result.append(np.float32(math.cos(2**i * math.pi * x)))
    return result
```

```python
df = df_dict_30fps[list(df_dict_30fps.keys())[0]]

# Extract unique joint names from column names while preserving order
joints = []
for col in df.columns:
    joint = '_'.join(col.split('_')[:-1])
    if joint not in joints:
        joints.append(joint)
print(joints)
```

```python
import pandas as pd

# Define a list of d_model values to test
#d_models = [1, 2, 3, 4]
N= 1

# Initialize an empty dictionary to store the encoded DataFrames
encoded_data_dict = {}

# Apply the encoding process for each DataFrame in the original dictionary
for key, df in df_dict_30fps.items():
    print(key)
    encoded_df = pd.DataFrame()
    for joint in joints:
        for coord_suffix in ['X', 'Y', 'Z']:
            coord_col = f'{joint}_{coord_suffix}position'
            encoded_col = f'{joint}_{coord_suffix}position_enc_d{N}'
            encoded_df[encoded_col] = df.apply(lambda row: pn(row[coord_col], N), axis=1)
    encoded_data_dict[key] = encoded_df
print('Done !')
```
