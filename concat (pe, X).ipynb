```python
import os
import numpy as np
import pandas as pd
```

```python
def get_data(folder_path):

    # Create an empty dictionary to store the DataFrames
    dataframes_dict = {}

    # Iterate over the .npy files in the folder
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.npy'):
            file_path = os.path.join(folder_path, file_name)

            # Load the numpy array from the file
            data_array = np.load(file_path)

            # Convert the numpy array to a DataFrame
            df = pd.DataFrame(data_array)

            # Use the file name (without extension) as the key in the dictionary
            key = os.path.splitext(file_name)[0]
            dataframes_dict[key] = df

    return dataframes_dict
```

```python
# Define the folder path containing the .npy files
x_dict = get_data('D:\Thèse_Toufik\\30fps_p')  # joint positions normilized
pe_dict = get_data('D:\Thèse_Toufik\\30fps_p_pes0')  # positional encoding of joint positions with N=0
```

Concatenate X and pe(X)

```python
# Initialize an empty dictionary to store the concatenated results
result_dict = {}

# Iterate over the keys of the dictionaries
for key in x_dict:
    print(key)
    if key in pe_dict:  # Ensure both dictionaries have the same keys
        # Concatenate DataFrames along columns
        print('x: ', len(x_dict[key]))
        print('pe(x): ', len(pe_dict[key]))
        result_dict[key] = pd.concat([x_dict[key], pe_dict[key]], axis=1)
```

save into npy file

```python
import numpy as np

# Create a directory to store the npy files (optional)
output_directory = 'npy_pose_pes0_concat'

# Create the directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

print('Joint position features with n=0 concat: ')
for key, df in result_dict.items():
    print(key)

    # Save the normalized DataFrame as a .npy file
    output_filepath = os.path.join(output_directory, f"{key}.npy")
    np.save(output_filepath, df.to_numpy())
    
    d=np.load(output_filepath)
    print(d.shape)
```
